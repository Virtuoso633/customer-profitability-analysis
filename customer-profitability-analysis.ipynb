# Q: identify the top 3 most profitable products `{Profitable products = Total sales - cost price}` , but as we do not know the cost price , so I am calculating top 3 products which contributes highest to revenue

# SOLUTION:: So first I have to take any product and sum all the entries of the product via TOT_SALES column, and then arrange `NET_TOT_SALES` in descending to get the top 3 product .

```python
product_sales = df1.groupby(['PROD_NBR', 'PROD_NAME'])['TOT_SALES'].sum().reset_index(name='NET_TOT_SALES')
```

```python
product_sales
```

```python
product_sales_sorted = product_sales.sort_values(by=['NET_TOT_SALES'], ascending=False)
product_sales_sorted.head(3)
```

# Q: who our most Loyal customers are. Are they married or unmarried, working or retired, premium or budget buyers etc…

# Solution :: Our most loyal customer, are based on number of times they revisit a shop for purchase, which we can get via counting the repeating entries in `LYLTY_CARD_NBR` column and then i am asking how many customers are revisting the store more than `n` times in the year, then getting the list of all those unique customers , and then via there `LYLTY_CARD_NBR` getting in which category they belong

```python
n = 10  # Number of purchases in a year

# Find LYLTY_CARD_NBRs with more than 'n' purchases
duplicate_lyc = df1['LYLTY_CARD_NBR'].value_counts()[lambda x: x > n].index

# Get rows where LYLTY_CARD_NBR appears more than 'n' times
duplicate_rows = df1[df1['LYLTY_CARD_NBR'].isin(duplicate_lyc)]

duplicate_rows['LYLTY_CARD_NBR'].unique()
```

```python
# Assuming 'duplicate_rows' contains the DataFrame from your previous code
unique_lylty_card_nbrs = duplicate_rows['LYLTY_CARD_NBR'].unique()

# Filter df2 based on the unique LYLTY_CARD_NBRs
filtered_df2 = df2[df2['LYLTY_CARD_NBR'].isin(unique_lylty_card_nbrs)]

# Display the filtered DataFrame
filtered_df2
```
